```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Load dataset
df = pd.read_csv('/content/Car_sales.csv')  # Notice: /content/ is the default upload folder

print(df.head())
```

```python
model_sales = df.groupby('Model')['Sales_in_thousands'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,7))
model_sales.plot(kind='bar', color='mediumseagreen', edgecolor='black')
plt.title('Car Model Sales', fontsize=16)
plt.xlabel('Car Model', fontsize=12)
plt.ylabel('Total Sales (in Thousands)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
model_sales_filtered = model_sales[model_sales > 5]
```

```python
fig = px.pie(
    names=model_sales_filtered.index,
    values=model_sales_filtered.values,
    title='Sales Distribution by Car Model (Top Models Only)',
    hole=0.3
)
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
fig = px.pie(
    names=model_sales.index,
    values=model_sales.values,
    title='Sales Distribution by Car Model'
)
fig.show()
```
