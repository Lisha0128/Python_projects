```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('GlobalLandTemperaturesByMajorCity.csv')

# Convert 'dt' to datetime
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

# Drop rows with missing temperature values
df = df.dropna(subset=['AverageTemperature'])

# Check the available cities
print(df['City'].unique())  # Print all cities to confirm
```

```python
# Filter for a city that definitely has enough data
city_data = df[df['City'] == 'New York']

# Check if data exists
print(city_data.head())
```

```python
# Sort by date
city_data = city_data.sort_values('dt')
```

```python
# Plotting Line Chart
plt.figure(figsize=(12,6))
plt.plot(city_data['dt'], city_data['AverageTemperature'], color='blue', linewidth=2)
plt.title('Temperature Trend Over Time (New York)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Temperature (°C)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Plotting Scatter Plot (Yearly Averages)
city_data['Year'] = city_data['dt'].dt.year
yearly_avg = city_data.groupby('Year')['AverageTemperature'].mean()

plt.figure(figsize=(12,6))
sns.scatterplot(x=yearly_avg.index, y=yearly_avg.values, color='red')
plt.title('Yearly Average Temperature (New York)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Temperature (°C)', fontsize=14)
plt.grid(True)
plt.show()
```
